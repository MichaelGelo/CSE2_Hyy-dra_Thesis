```python
import numpy as np
import pynq
import pynq.lib.dma
import time
from pynq import Xlnk

ref_fname = "ref2_1M.fasta" # <--- Reference FASTA file name
que_fname = "que2_128.fasta" # <--- Query FASTA file name

dnaref = "" # Sequence for the reference
dnaque = "" # Sequence for the query

# --- Function to Read FASTA Sequence ---
def read_fasta(file_path):
    """Reads a FASTA file, concatenating all sequence lines and skipping headers."""
    sequence = ""
    try:
        with open(file_path, 'r') as f:
            for line in f:
                line = line.strip()
                if line.startswith('>'):
                    # Skip the FASTA header line
                    pass
                else:
                    # Append sequence data
                    sequence += line
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        # Return an empty string if the file doesn't exist
        return ""
    
    return sequence

# --- Load Sequences ---
print(f"Loading reference sequence from: {ref_fname}")
dnaref = read_fasta(ref_fname)
print(f"Reference sequence loaded. Length: **{len(dnaref)}**")
print("-" * 30)

print(f"Loading query sequence from: {que_fname}")
dnaque = read_fasta(que_fname)
print(f"Query sequence loaded. Length: **{len(dnaque)}**")
print("-" * 30)

```

```python
ol = pynq.Overlay("DMA_wrapper.bit")

hyyro_ip = ol.hyyro_0
dma0 = ol.axi_dma_0   # will be used for reference send and zero-index recv
dma1 = ol.axi_dma_1   # will be used for query send and score recv

query_str = dnaque
ref_str = dnaref

m = len(query_str)
n = len(ref_str)


```

```python
xlnk = Xlnk()

xlnk.xlnk_reset()
ref_buffer = xlnk.cma_array(shape=(n,), dtype=np.uint8)
query_buffer = xlnk.cma_array(shape=(m,), dtype=np.uint8)

ref_buffer[:] = np.frombuffer(ref_str.encode('ascii'), dtype=np.uint8)
query_buffer[:] = np.frombuffer(query_str.encode('ascii'), dtype=np.uint8)

zero_idx_buffer = xlnk.cma_array(shape=(1024,), dtype=np.uint32)
score_buffer = xlnk.cma_array(shape=(32,), dtype=np.uint32)

```

```python
zero_idx_buffer.fill(-1)
score_buffer.fill(-1)

def print_ctrl_status(ip):
    ctrl = ip.read(0x00)
    ap_start = bool(ctrl & 0x01)
    ap_done  = bool(ctrl & 0x02)
    ap_idle  = bool(ctrl & 0x04)
    ap_ready = bool(ctrl & 0x08) if (ctrl & 0x08) else False
    print(f"control reg 0x00 = 0x{ctrl:02x}, ap_start={ap_start}, ap_done={ap_done}, ap_idle={ap_idle}, ap_ready={ap_ready}")
    return ctrl

# helper to check `halted` attribute safely
def channel_halted(dma_recv):
    # attribute exists in PYNQ DMA driver as used earlier
    return getattr(dma_recv, "halted", True)

# clear previous activity best-effort
import time
from math import ceil

# small helper, reuse your channel_halted function and print_ctrl_status if available
def stop_all_channels():
    try:
        if hasattr(dma0.recvchannel, "stop"): dma0.recvchannel.stop()
    except Exception as e:
        print("stop dma0.recv err", e)
    try:
        if hasattr(dma1.recvchannel, "stop"): dma1.recvchannel.stop()
    except Exception as e:
        print("stop dma1.recv err", e)
    try:
        if hasattr(dma0.sendchannel, "stop"): dma0.sendchannel.stop()
    except Exception as e:
        print("stop dma0.send err", e)
    try:
        if hasattr(dma1.sendchannel, "stop"): dma1.sendchannel.stop()
    except Exception as e:
        print("stop dma1.send err", e)
    time.sleep(0.005)

def try_recv_size(size_bytes, send_timeout=5.0, recv_timeout=2.0):
    """
    Try one probe where we tell S2MM to expect size_bytes.
    Returns (success_bool, actual_bytes_arrived, zero_preview, score_preview)
    """
    print(f"\n=== TRYING recv size = {size_bytes} bytes ===")
    stop_all_channels()

    # clear buffers to known sentinel
    zero_idx_buffer.fill(0xFFFFFFFF)
    score_buffer.fill(0xFFFFFFFF)

    # arm recv with explicit nbytes
    try:
        dma0.recvchannel.transfer(zero_idx_buffer, 0, size_bytes)
        dma1.recvchannel.transfer(score_buffer, 0, score_buffer.nbytes)  # keep same
    except Exception as e:
        print("Error arming recv:", e)
        return False, 0, None, None

    # arm sends
    try:
        dma0.sendchannel.transfer(ref_buffer, 0, ref_buffer.nbytes)
        dma1.sendchannel.transfer(query_buffer, 0, query_buffer.nbytes)
    except Exception as e:
        print("Error arming send:", e)
        # stop recv so we don't leak
        try:
            dma0.recvchannel.stop()
        except:
            pass
        return False, 0, None, None

    # write kernel args and start
    hyyro_ip.write(0x10, m)
    hyyro_ip.write(0x18, n)
    hyyro_ip.write(0x00, 1)

    # wait for sends to finish (poll)
    t0 = time.time()
    while True:
        sended = channel_halted(dma0.sendchannel) and channel_halted(dma1.sendchannel)
        if sended:
            break
        if time.time() - t0 > send_timeout:
            print("Timeout waiting for send channels to halt (probe).")
            break
        time.sleep(0.001)

    # now poll recv for a (short) time
    t1 = time.time()
    got = False
    while True:
        rcvd = channel_halted(dma0.recvchannel) and channel_halted(dma1.recvchannel)
        if rcvd:
            got = True
            break
        if time.time() - t1 > recv_timeout:
            # timed out
            break
        time.sleep(0.001)

    # measure what arrived in zero_idx_buffer: we'll interpret up to size_bytes/4 words
    words = int(ceil(size_bytes / 4.0))
    preview = zero_idx_buffer[:min(32, words)].tolist()
    preview_scores = score_buffer[:min(8, score_buffer.size)].tolist()

    if got:
        print("RECV HALTED within timeout.")
        # try to determine how many valid words arrived by scanning for 0xFFFFFFFF sentinel
        valid_count = 0
        for i in range(words):
            val = int(zero_idx_buffer[i])
            if val == 0xFFFFFFFF:
                # sentinel or unwritten region
                break
            valid_count += 1
        print(f"Preview zero_idx (first {min(32, words)}): {preview}")
        print(f"Detected {valid_count} non-0xFFFFFFFF entries (estimated).")
        print(f"Score preview (first 8 words): {preview_scores}")
        return True, valid_count * 4, preview, preview_scores
    else:
        print("RECV did NOT halt within timeout.")
        print(f"Preview zero_idx (first {min(32, words)}): {preview}")
        print(f"Score preview (first 8 words): {preview_scores}")
        # try best-effort stop to recover
        try:
            dma0.recvchannel.stop()
        except Exception:
            pass
        return False, 0, preview, preview_scores

# Probe list (start very small and grow)
probe_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096]
result = None
for s in probe_sizes:
    succ, arrived_bytes, preview, sprev = try_recv_size(s, send_timeout=3.0, recv_timeout=1.0)
    if succ:
        print(f"SUCCESS at size {s}, arrived_bytes estimate = {arrived_bytes}")
        result = (s, arrived_bytes, preview, sprev)
        break

if result is None:
    print("\nNo probe size finished within timeouts. Try increasing recv_timeout or check HLS for input-side blocking.")
else:
    print("\nProbe result:", result)
import time
from math import ceil

# small helper, reuse your channel_halted function and print_ctrl_status if available
def stop_all_channels():
    try:
        if hasattr(dma0.recvchannel, "stop"): dma0.recvchannel.stop()
    except Exception as e:
        print("stop dma0.recv err", e)
    try:
        if hasattr(dma1.recvchannel, "stop"): dma1.recvchannel.stop()
    except Exception as e:
        print("stop dma1.recv err", e)
    try:
        if hasattr(dma0.sendchannel, "stop"): dma0.sendchannel.stop()
    except Exception as e:
        print("stop dma0.send err", e)
    try:
        if hasattr(dma1.sendchannel, "stop"): dma1.sendchannel.stop()
    except Exception as e:
        print("stop dma1.send err", e)
    time.sleep(0.005)

def try_recv_size(size_bytes, send_timeout=5.0, recv_timeout=2.0):
    """
    Try one probe where we tell S2MM to expect size_bytes.
    Returns (success_bool, actual_bytes_arrived, zero_preview, score_preview)
    """
    print(f"\n=== TRYING recv size = {size_bytes} bytes ===")
    stop_all_channels()

    # clear buffers to known sentinel
    zero_idx_buffer.fill(0xFFFFFFFF)
    score_buffer.fill(0xFFFFFFFF)

    # arm recv with explicit nbytes
    try:
        dma0.recvchannel.transfer(zero_idx_buffer, 0, size_bytes)
        dma1.recvchannel.transfer(score_buffer, 0, score_buffer.nbytes)  # keep same
    except Exception as e:
        print("Error arming recv:", e)
        return False, 0, None, None

    # arm sends
    try:
        dma0.sendchannel.transfer(ref_buffer, 0, ref_buffer.nbytes)
        dma1.sendchannel.transfer(query_buffer, 0, query_buffer.nbytes)
    except Exception as e:
        print("Error arming send:", e)
        # stop recv so we don't leak
        try:
            dma0.recvchannel.stop()
        except:
            pass
        return False, 0, None, None

    # write kernel args and start
    hyyro_ip.write(0x10, m)
    hyyro_ip.write(0x18, n)
    hyyro_ip.write(0x00, 1)

    # wait for sends to finish (poll)
    t0 = time.time()
    while True:
        sended = channel_halted(dma0.sendchannel) and channel_halted(dma1.sendchannel)
        if sended:
            break
        if time.time() - t0 > send_timeout:
            print("Timeout waiting for send channels to halt (probe).")
            break
        time.sleep(0.001)

    # now poll recv for a (short) time
    t1 = time.time()
    got = False
    while True:
        rcvd = channel_halted(dma0.recvchannel) and channel_halted(dma1.recvchannel)
        if rcvd:
            got = True
            break
        if time.time() - t1 > recv_timeout:
            # timed out
            break
        time.sleep(0.001)

    # measure what arrived in zero_idx_buffer: we'll interpret up to size_bytes/4 words
    words = int(ceil(size_bytes / 4.0))
    preview = zero_idx_buffer[:min(32, words)].tolist()
    preview_scores = score_buffer[:min(8, score_buffer.size)].tolist()

    if got:
        print("RECV HALTED within timeout.")
        # try to determine how many valid words arrived by scanning for 0xFFFFFFFF sentinel
        valid_count = 0
        for i in range(words):
            val = int(zero_idx_buffer[i])
            if val == 0xFFFFFFFF:
                # sentinel or unwritten region
                break
            valid_count += 1
        print(f"Preview zero_idx (first {min(32, words)}): {preview}")
        print(f"Detected {valid_count} non-0xFFFFFFFF entries (estimated).")
        print(f"Score preview (first 8 words): {preview_scores}")
        return True, valid_count * 4, preview, preview_scores
    else:
        print("RECV did NOT halt within timeout.")
        print(f"Preview zero_idx (first {min(32, words)}): {preview}")
        print(f"Score preview (first 8 words): {preview_scores}")
        # try best-effort stop to recover
        try:
            dma0.recvchannel.stop()
        except Exception:
            pass
        return False, 0, preview, preview_scores

# Probe list (start very small and grow)
probe_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096]
result = None
for s in probe_sizes:
    succ, arrived_bytes, preview, sprev = try_recv_size(s, send_timeout=3.0, recv_timeout=1.0)
    if succ:
        print(f"SUCCESS at size {s}, arrived_bytes estimate = {arrived_bytes}")
        result = (s, arrived_bytes, preview, sprev)
        break

if result is None:
    print("\nNo probe size finished within timeouts. Try increasing recv_timeout or check HLS for input-side blocking.")
else:
    print("\nProbe result:", result)
```
